```python
li_location=[]
df1=pd.DataFrame({"tweet":[],"date":[]})
df_temp=pd.DataFrame()
li=["#JawaharlalNehru"]
temp=0
for i in li:
    for j in li_state:
        c = twint.Config()
        #Custom_csv = ["tweet","date"]
        c.Search = i
        c.Lang="en"
        c.Limit = 20
        c.near= j[1],j[2],100
        c.Pandas = True
        twint.run.Search(c)
        df_temp=df_temp.append(twint.storage.panda.Tweets_df)
        for k in range(df_temp.shape[0]-df1.shape[0]):
            li_location.append(j[0])
        #df1=df1.append(twint.storage.panda.Tweets_df) 
        df1=df1.append(twint.storage.panda.Tweets_df)
        
```

```python
df1["state"]=li_location
df1=df1.reset_index()
df1.drop("index",axis=1,inplace=True)
df1 = df1.drop(df1[df1.language != "en"].index)
df1=df1.reset_index()
df1.drop("index",axis=1,inplace=True)
```

```python
new_dfs=pd.DataFrame({"tweet":[]})
li_sentiment=[]
for i in range(df1.shape[0]):   
    x=re.sub('#\S+',' ',str(df1.loc[i,"tweet"]))
    x=re.sub('@\S+',' ',x)
    x=re.sub('\S+@\S+',' ',x)
    x=re.sub('http\S+',' ',x)
    x=re.sub('[^A-Za-z]',' ',x)
    x=re.sub('\d+',' ',x)
    li=[]
    for l in x.lower().split():
        if l.isalpha()==True:
            li.append(l)
    str2=""
    for l in li:
        if l not in stop_li:
                str2+=l+" "
    str_temp = word_tokenize(str2.lower())
    str3=""
    for k in str_temp:
        
        str3+=lemmatizer.lemmatize(k)+" "
        
    new_dfs.at[i,"tweet"]=str3.strip()
    #new_dfs.at[i,"date"]=df1.loc[i,"date"]
```

```python
from nltk.tokenize import word_tokenize
import itertools
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences
import codecs
from tqdm import tqdm
import numpy as np
li_train1=[]
for i in new_dfs["tweet"]:
    words=word_tokenize(i)
    li_train1.append(words)
li_train = list(itertools.chain(*li_train1))
li_all=li_train
li_uniquetrain=set(li_train)
li_unique=set(li_train)
tokenizer = Tokenizer(num_words=len(li_unique), lower=True, char_level=False)
tokenizer.fit_on_texts(li_all)
word_seq_train = tokenizer.texts_to_sequences(li_train1)
word_index = tokenizer.word_index
li_all1=li_train1
max=0
for i in li_all1:
    if len(i)>max:
        max=len(i)
train=pad_sequences(word_seq_train, maxlen=100)
```

```python
from keras.models import load_model
model = load_model('lstm_sentiment1.h5')
predictions = model.predict(train)
```

```python
final=list([list(x) for x in predictions])
y_prediction=[]
for x in final:
    if x[0]>x[1]:
        y_prediction.append(0)
    elif x[0]<x[1]:
        y_prediction.append(1)
```

```python
li_location=set(df["State"].values)
li_positive=[]
li_negative=[]
for i in df_state["State"].values:
    pos=0
    neg=0
    for j in range(df.shape[0]):
        if df.loc[j,"State"]==i and df.loc[j,"sentiment"]==1:
            pos+=1
        elif df.loc[j,"State"]==i and df.loc[j,"sentiment"]==0:
            neg+=1
    li_positive.append(pos)
    li_negative.append(neg)
li_positive_p=[]
for i in range(len(li_positive)):
    if li_positive[i]+li_negative[i]!=0:
        li_positive_p.append(li_positive[i]/(li_positive[i]+li_negative[i])*100)
    else:
        li_positive_p.append("na")
li_negative_p=[]
for i in range(len(li_positive)):
    if li_positive[i]+li_negative[i]!=0:
        li_negative_p.append(li_negative[i]/(li_positive[i]+li_negative[i])*100)
    else:
        li_negative_p.append("na")
li_party=set(df["Party"].values)
li_positive_party=[]
li_negative_party=[]
for i in li_party:
    pos_p=0
    neg_p=0
    for j in range(df.shape[0]):
        if df.loc[j,"Party"]==i and df.loc[j,"sentiment"]==1:
            pos_p+=1
        elif df.loc[j,"Party"]==i and df.loc[j,"sentiment"]==0:
            neg_p+=1
    li_positive_party.append(pos_p)
    li_negative_party.append(neg_p)
li_positive_partyp=[]
for i in range(len(li_positive_party)):
    if li_positive_party[i]+li_negative_party[i]==0:
        li_positive_partyp.append("na")
    else:
        li_positive_partyp.append(li_positive_party[i]/(li_positive_party[i]+li_negative_party[i])*100)
li_negative_partyp=[]
for i in range(len(li_positive_party)):
    if li_positive_party[i]+li_negative_party[i]==0:
        li_negative_partyp.append("na")
    else:
        li_negative_partyp.append(li_negative_party[i]/(li_positive_party[i]+li_negative_party[i])*100)
```

```python
df2=pd.DataFrame({"State":[],"Party":[],"Positive":[],"Negative":[]})
df3=pd.DataFrame({"Party":[],"Positive":[],"Negative":[]})
df2["State"]=df_state["State"]
df2["Party"]=df_state["Party"]
df2["Positive"]=li_positive_p
df2["Negative"]=li_negative_p
df3["Party"]=list(li_party)
df3["Positive"]=li_positive_partyp
df3["Negative"]=li_negative_partyp
```
